Exercise 1

Given the code below:


```python
text = 'Business Analytics'
```

What will the following expression produce?

```python
text[4]
text[2:4]
text[:4]
text[4:]
text[-2]
text[-4:-2]
text[-2:]
text[:-2]
```

Answer:

```python
text = "Business Analytics"
print(text[4])
print(text[2:4])
print(text[4:])
print(text[:4])
print(text[-2])
print(text[-4:-2])
print(text[-2:])
print(text[:-2])
```

Given the code below:

```python
phrase = 'One Ring to rule them all'
```

What string slicing expression gives out the following words:

```python
'One '
'Ring'
'all'
'them'
'One Ring to rule them all'
```

Answer:

```python
phrase = "One ring to rule them all"
print(phrase[:4])
print(phrase[4:8])
print(phrase[-3:])
print(phrase[17:-4])
print(phrase[:])
```

CHALLENGE

Write a program which repeatedly reads a full name, then does the following:

strips the name of any trailing spaces capitalizes the first letter puts the rest into smaller letters

```python
for i in range(10) :
  name = input("Enter your full name: ")
  if name == "exit" :
      break
  print(name.strip().capitalize())
```
